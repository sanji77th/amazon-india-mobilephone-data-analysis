## Load the dataset

```python
df = pd.read_csv('data/raw_data.csv')
df.head()
```

### Clean title row

```python
product = []
for data_num in range(len(df['title'])):
    data_lis = df['title'][data_num].split('(')
    product.append(data_lis[0])
    if len(data_lis[0])>40:
        print(data_lis[0],data_num)

product[5] = "Samsung Galaxy M04"
product[27] = "Nokia C22"
product[36] = "Nokia 2660"
print(product) 

df["product"] = pd.Series(product)
```
